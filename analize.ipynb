```python
data ['Year']=pd.to_datetime(data['Year'], format='%Y')
data.info() 
```

```python
data['Genre'].value_counts()
```

```python
sns.countplot(x='Genre',data=data,order=data['Genre'].value_counts().index)
plt.xticks(rotation=90); 
```

```python
numeric_colums = data.select_dtypes(include=['float64', 'int64'])

correlation = numeric_colums.corr()

plt.figure(figsize=(10,8))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title ('Pairwise Correlation of Video Game Sales Features')
plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.lineplot(data=data, x='Year', y='Genre', hue='Genre', palette='tab10')

plt.xlabel('Year')
plt.ylabel('Sales (in millions)')
plt.title('Video Game Sales Over the Years')
plt.legend(title='Sales Category', loc='upper left')
plt.grid(True)
plt.show()

```

```python
area_sales = data[['NA_Sales','EU_Sales', 'JP_Sales', 'Other_Sales']]
area_sales = area_sales.melt(var_name='Area',value_name='Total_Sales')
area_sales = area_sales.groupby('Area')['Total_Sales'].sum().reset_index()

plt.figure(figsize=(10,5))
plt.pie(area_sales['Total_Sales'],labels=area_sales['Area'],colors=sns.color_palette('magma_r'),autopct='%.2f%%')
plt.show()
```

```python
#top 10 publishers
top_publishers = data['Publisher'].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_publishers.values,y=top_publishers.index,palette='viridis')
plt.title('Top 10 Publishers')
plt.xlabel('Number of Games Published')
plt.ylabel('Publisher')
plt.show()
```

```python
#botton 10 publishers

bottom_publishers = data['Publisher'].value_counts().tail(10)
plt.figure(figsize=(10,5))
sns.barplot(x=bottom_publishers.values,y=bottom_publishers.index,palette='viridis')
plt.title('Bottom 10 Publishers')
plt.xlabel('Number of Games Published')
plt.ylabel('Publisher')
plt.show()
```

```python
# all sales JP 
jp_sales = data.groupby('Year')['JP_Sales'].sum().reset_index()
plt.figure(figsize=(12,6))
sns.lineplot(data=jp_sales,x='Year',y='JP_Sales',color='purple')
plt.title('JP Sales Over the Years')
plt.xlabel('Year')
plt.ylabel('Sales (in millions)')
plt.grid(True)
plt.show()
```

```python
#all sales NA

na_sales = data.groupby('Year')['NA_Sales'].sum().reset_index()
plt.figure(figsize=(12,6))
sns.lineplot(data=na_sales,x='Year',y='NA_Sales',color='purple')
plt.title('NA Sales Over the Years')
plt.xlabel('Year')
plt.ylabel('Sales (in millions)')
plt.grid(True)
plt.show()
```

```python
# all sales EU

eu_sales = data.groupby('Year')['EU_Sales'].sum().reset_index()
plt.figure(figsize=(12,6))
sns.lineplot(data=eu_sales,x='Year',y='EU_Sales',color='purple')
plt.title('EU Sales Over the Years')
plt.xlabel('Year')
plt.ylabel('Sales (in millions)')
plt.grid(True)
plt.show()
```

```python
# all sales Other

other_sales = data.groupby('Year')['Other_Sales'].sum().reset_index()
plt.figure(figsize=(12,6))
sns.lineplot(data=other_sales,x='Year',y='Other_Sales',color='purple')
plt.title('Other Sales Over the Years')
plt.xlabel('Year')
plt.ylabel('Sales (in millions)')
plt.grid(True)
plt.show()
```
